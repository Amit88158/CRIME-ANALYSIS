```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

**# -------------------- DATA CLEANING --------------------**#


```python

df = pd.read_csv("crime.csv")
print(df.head())
```

```python
df.info()
```

**# Handle missing values**


```python
df['HOUR'].fillna(df['HOUR'].median(), inplace=True)
df['MINUTE'].fillna(df['MINUTE'].median(), inplace=True)
df['NEIGHBOURHOOD'].fillna("Unknown", inplace=True)
df['HUNDRED_BLOCK'].fillna("Unknown", inplace=True)
df.info()
```

# Convert datatypes

```python
df['HOUR'] = df['HOUR'].astype(int)
df['MINUTE'] = df['MINUTE'].astype(int)
```

```python
df['DATETIME'] = pd.to_datetime(df[['YEAR','MONTH','DAY','HOUR','MINUTE']])
sns.set_style("whitegrid")
```

# -------------------- Z-TEST --------------------
# Example: Compare sample mean (HOUR) with population mean

```python
sample = df_clean['HOUR'].sample(1000, random_state=42)
pop_mean = df_clean['HOUR'].mean()
pop_std = df_clean['HOUR'].std()

z_stat = (sample.mean() - pop_mean) / (pop_std / np.sqrt(len(sample)))

p_val_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("\nZ-Test:")
print("Z-Statistic:", z_stat)
print("P-Value:", p_val_z)
```
